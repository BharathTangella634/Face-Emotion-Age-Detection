```python
import pandas as pd
df = pd.read_csv('csv_dataset/utkface_dataset.csv')

df.head()
```

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Root directory
root_dir = "/mnt/8b4bbd12-99b7-4ef1-9218-be56afd51a3d/Facial Emotion and Age Prediction/Facial_Age_estimation_PyTorch/utkcropped"

# Randomly sample 2 rows
sample_df = df.sample(2)

plt.figure(figsize=(10,5))

for i, (_, row) in enumerate(sample_df.iterrows()):
    
    img_path = os.path.join(root_dir, row['image_name'])
    
    # Load image
    img = Image.open(img_path).convert("RGB")
    
    # Convert to array to get shape
    img_array = np.array(img)
    
    plt.subplot(1, 2, i+1)
    plt.imshow(img_array)
    plt.axis("off")
    
    plt.title(f"Shape: {img_array.shape}\nAge: {row['age']} | Ethnicity: {row['ethnicity']}")

plt.tight_layout()
plt.show()
```
